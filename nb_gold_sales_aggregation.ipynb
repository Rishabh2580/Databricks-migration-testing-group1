```python
%sql
CREATE OR REPLACE TABLE
capstone_db.gold_sales_summary AS SELECT
category,ROUND(SUM(try_cast(sales AS DOUBLE)),2) AS total_sales,
COUNT(DISTINCT order_id) AS order_count
FROM
capstone_db.silver_superstore_orders
GROUP BY category
ORDER BY category
```
