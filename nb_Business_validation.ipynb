```python
%sql
SELECT category, ROUND(SUM(TRY_CAST(sales AS DOUBLE)),2) AS total_sales
FROM capstone_db.silver_superstore_orders
GROUP BY category
ORDER BY category
```

```python
%sql
SELECT
region,
ROUND(SUM(TRY_CAST(sales AS DOUBLE)),2) AS total_sales
FROM capstone_db.silver_superstore_orders
GROUP BY region
ORDER BY region
```

```python
%sql
SELECT
  g.category,
  g.total_sales AS databricks_sales,
  a.total_sales AS alteryx_sales,
  g.order_count AS databricks_orders,
  a.order_count AS alteryx_orders
FROM capstone_db.gold_sales_summary g
JOIN capstone_db.category_sales_summary a
  ON g.category = a.category
```
